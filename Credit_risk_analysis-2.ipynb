```python
import pandas as pd
import numpy as np

# Load main table
app_train = pd.read_csv('application_train.csv')
app_test = pd.read_csv('application_test.csv')


bureau = pd.read_csv('bureau.csv')
# bureau_balance = pd.read_csv('bureau_balance.csv')
# prev = pd.read_csv('previous_application.csv')
# inst = pd.read_csv('installments_payments.csv')
# pos = pd.read_csv('POS_CASH_balance.csv')
# credit = pd.read_csv('credit_card_balance.csv')

print(f"Train shape: {app_train.shape}, Test shape: {app_test.shape}")
```

```python
# Target distribution
print(app_train['TARGET'].value_counts(normalize=True))

# Fill missing
app_train = app_train.fillna(-999)
app_test = app_test.fillna(-999)

# Example ratio features
app_train['CREDIT_INCOME_RATIO'] = app_train['AMT_CREDIT'] / (app_train['AMT_INCOME_TOTAL'] + 1)
app_train['ANNUITY_INCOME_RATIO'] = app_train['AMT_ANNUITY'] / (app_train['AMT_INCOME_TOTAL'] + 1)

# Apply the same ratio features to app_test
app_test['CREDIT_INCOME_RATIO'] = app_test['AMT_CREDIT'] / (app_test['AMT_INCOME_TOTAL'] + 1)
app_test['ANNUITY_INCOME_RATIO'] = app_test['AMT_ANNUITY'] / (app_test['AMT_INCOME_TOTAL'] + 1)


# Label encoding categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in app_train.select_dtypes(include='object').columns:
    app_train[col] = le.fit_transform(app_train[col].astype(str))
    app_test[col] = le.transform(app_test[col].astype(str))

# TODO: You can write merge functions to pull aggregated features from other tables (e.g., mean of overdue days from bureau)
```

```python
from sklearn.model_selection import train_test_split
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score

X = app_train.drop(columns=['TARGET', 'SK_ID_CURR'])
y = app_train['TARGET']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

model = LGBMClassifier(n_estimators=500, max_depth=8, learning_rate=0.03)
model.fit(X_train, y_train)

y_pred_proba = model.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, y_pred_proba)
print(f"Validation AUC: {auc:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = model.predict(X_val)
print(confusion_matrix(y_val, y_pred))
print(classification_report(y_val, y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# Define the model
xgb_model = XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.05,
    # subsample=0.8,
    # colsample_bytree=0.8,
    use_label_encoder=False,
    eval_metric='auc',
    random_state=42
)

# Train
xgb_model.fit(X_train, y_train)

# Predict
xgb_preds = xgb_model.predict_proba(X_val)[:, 1]

# Evaluate
xgb_auc = roc_auc_score(y_val, xgb_preds)
print(f"XGBoost Validation AUC: {xgb_auc:.4f}")
```

```python
lgbm_preds = model.predict_proba(X_val)[:, 1]
lgbm_auc = roc_auc_score(y_val, lgbm_preds)

print(f"LightGBM AUC:  {lgbm_auc:.4f}")
print(f"XGBoost AUC:   {xgb_auc:.4f}")
```

```python
# Copy original DataFrame to avoid mutation
app = application_train.copy()
y = app["TARGET"]

# Drop target from features
app.drop(columns=["TARGET"], inplace=True)

# Optional: drop obvious ID columns
app.drop(columns=["SK_ID_CURR"], inplace=True)
```

```python
# Aggregate bureau-level features
bureau_agg = bureau.groupby("SK_ID_CURR").agg({
    "CREDIT_DAY_OVERDUE": ["mean", "max"],
    "AMT_CREDIT_SUM": "sum",
    "AMT_CREDIT_SUM_DEBT": "sum"
})
bureau_agg.columns = ['BUREAU_OVERDUE_MEAN', 'BUREAU_OVERDUE_MAX', 'BUREAU_CREDIT_SUM', 'BUREAU_DEBT_SUM']
bureau_agg.reset_index(inplace=True)

# Merge with app
app = application_train.merge(bureau_agg, on="SK_ID_CURR", how="left")
```

```python
# Combine with engineered features
X = app.drop(columns=["TARGET"])

# Fix object columns: label encode categorical features
for col in X.columns:
    if X[col].dtype == 'object':
        X[col] = X[col].astype('category').cat.codes
```

```python
from sklearn.preprocessing import LabelEncoder

# Make a copy just to be safe
X_encoded = X.copy()

# Encode all object (categorical) columns
for col in X_encoded.columns:
    if X_encoded[col].dtype == 'object':
        # Fill missing with mode
        X_encoded[col] = X_encoded[col].fillna(X_encoded[col].mode()[0])

        # Use LabelEncoder for consistent numeric conversion
        le = LabelEncoder()
        X_encoded[col] = le.fit_transform(X_encoded[col])
```

```python
# Reassign X and y now that app has engineered features
y = application_train["TARGET"]
X = app.drop(columns=["TARGET", "SK_ID_CURR"])  # if SK_ID_CURR is still present
```

```python
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
X.head()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import roc_auc_score
import lightgbm as lgb
import xgboost as xgb
```

```python
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_idx, val_idx in sss.split(X, y):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
```

```python
import lightgbm as lgb
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import LabelEncoder

# The preprocessing is now done before the split, so no need to do it here.

# Fix object columns: label encode categorical features in the split data
for col in X_train.columns:
    if X_train[col].dtype == 'object':
        le = LabelEncoder()
        # Fit on train and transform train and val to ensure consistency
        X_train[col] = le.fit_transform(X_train[col].astype(str))
        X_val[col] = le.transform(X_val[col].astype(str))


lgbm_model = lgb.LGBMClassifier(
    objective='binary',
    n_estimators=500,
    learning_rate=0.05,
    max_depth=7,
    random_state=42
)

lgbm_model.fit(X_train, y_train)
y_pred_lgb = lgbm_model.predict_proba(X_val)[:, 1]
auc_lgb = roc_auc_score(y_val, y_pred_lgb)
print(f"LightGBM AUC-ROC: {auc_lgb:.4f}")
```

```python
xgb_model = xgb.XGBClassifier(
    objective='binary:logistic',
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=7,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict_proba(X_val)[:, 1]
auc_xgb = roc_auc_score(y_val, y_pred_xgb)
print(f"XGBoost AUC-ROC: {auc_xgb:.4f}")
```

```python
import shap
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Ensure X_val has the same encoding as X_train
X_val_encoded = X_val.copy()
for col in X_val_encoded.columns:
    if X_val_encoded[col].dtype == 'object':
        le = LabelEncoder()
        # Fit on X_train and transform X_val to ensure consistency
        le.fit(X_train[col].astype(str))
        X_val_encoded[col] = le.transform(X_val_encoded[col].astype(str))


# SHAP prefers TreeExplainer for tree models
explainer_lgb = shap.Explainer(lgbm_model, X_val_encoded)
shap_values_lgb = explainer_lgb(X_val_encoded)

explainer_xgb = shap.Explainer(xgb_model, X_val_encoded)
shap_values_xgb = explainer_xgb(X_val_encoded)
```
